```python
import numpy as np
import pandas as pd
import os
```

```python
os.listdir('data')
```

```python
df_env = pd.read_csv('data/Env_QH.csv', delimiter=';', encoding='utf-8')
print(len(df_env))
df_env = df_env.rename(columns={'Temp': 'temp', 'RH': 'rh', 'Tgrad': 'tgrad', 'Patm': 'pressure', 'Pluvio':'pluvio'})
df_env.tail()
```

```python
df_AllPM = pd.read_csv('data/AllPM_QH.csv', delimiter=';')
df_AllPM = df_AllPM.rename(columns={'ref': 'PM_ref', '#6182': 'PM_6182', '#6179': 'PM_6179', '#617B': 'PM_617B', 'pm2.5#6182': 'PM25_6182', 'pm2.5#6179': 'PM25_6179', 'pm2.5#617B': 'PM25_617B'})
len(df_AllPM)
df_AllPM.tail()
len(np.unique(df_AllPM['PM25_617B']))
```

```python
df_NO2 = pd.read_csv('data/AllNO2_QH.csv', delimiter=";")
df_NO2 = df_NO2.rename(columns={'ref': 'NO2_ref', '#61FD': 'NO2_61FD', '#61F0': 'NO2_61F0', '#61EF': 'NO2_61EF'})
print(len(df_NO2))
df_NO2.head()
len(np.unique(df_NO2['NO2_61FD']))
```

```python
df_env = df_env.set_index('date').T
df_env.head()
```

```python
df_data = pd.concat([df_AllPM, df_NO2[['NO2_ref', 'NO2_61FD', 'NO2_61F0', 'NO2_61EF']]], axis=1)
```

```python
df_data.head()
```

```python
out_df = pd.DataFrame(columns=['date','PM_ref', 'PM_6182', 'PM_6179', 'PM_617B', 'PM25_6182',\
                               'PM25_6179', 'PM25_617B','NO2_ref', 'NO2_61FD', 'NO2_61F0',\
                               'NO2_61EF', 'temp', 'rh', 'tgrad', 'pressure', 'pluvio'])

for i in range(len(df_data)):
    row = df_data.iloc[i]
    date = row.date
    env = df_env[date]
    out_df.loc[i] = [date, row.PM_ref, row.PM_6182, row.PM_6179, row.PM_617B, row.PM25_6182,\
                     row.PM25_6179, row.PM25_617B, row.NO2_ref, row.NO2_61FD, row.NO2_61F0,\
                     row.NO2_61EF, env.temp, env.rh, env.tgrad, env.pressure, env.pluvio]
```

```python
out_df.head()
```

```python
out_df.to_pickle('data/micro_sud3.pkl')
```

```python
out_df = out_df[pd.notnull(out_df).all(axis=1)]
tmp_df = out_df[['PM_6182', 'PM_6179', 'PM_617B', 'PM25_6182',\
                 'PM25_6179', 'PM25_617B', 'NO2_61FD', 'NO2_61F0',\
                 'NO2_61EF', 'temp', 'rh', 'tgrad', 'pressure', 'pluvio']]
normalized_df = (tmp_df - tmp_df.mean()) / tmp_df.std()
out_df[['PM_6182', 'PM_6179', 'PM_617B', 'PM25_6182',\
         'PM25_6179', 'PM25_617B', 'NO2_61FD', 'NO2_61F0',\
         'NO2_61EF', 'temp', 'rh', 'tgrad', 'pressure', 'pluvio']] = normalized_df

out_df.to_pickle('data/micro_sud3_normalized.pkl')
print(len(np.unique(out_df['NO2_61FD'])))
```
